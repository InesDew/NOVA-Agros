Correlation coefficients quantify the strength and direction of the linear relationship between two variables and can range from -1 to 1. In our dataset,
* The "_quantity" columns are all correlated positivly exclusively which means that when a quantity increases, the other quantities will also increase. When a quantity decreases, the other quantities also decrease.

* The "_quantity" columns are correlated pretty high which means that the quantities have strong linear relationships to each other. Only the fish output quantity has a generally weaker linear relationship to the other quantities. 

* There are 2 pairs of quantities that are perfectly, positively linear correlated (corr = 1): (ag land quantity and cropland quantity) & (crop output quantity and output quantity) which means that if one variable increases, the other variable also increases in a linear fashion.
