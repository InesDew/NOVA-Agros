**Deep Dive: Norway, Brazil India**

To understand the agricultural production across the world better, we focused on three countries from different continents: Norway, Brazil and India. We focused on these countries as they all have high GDPs in comparison to other countries on the continents/regions (Norway in Europe, India in South Asia, Brazil in South America).

```python
# Method 4
obj.output_area_plot("Norway", True)
obj.output_area_plot("Brazil", True)
obj.output_area_plot("India", True)
```

The graphs show the quantity of Crop Output in blue (including grains, vegetables, fruits, and other plants), Fish Output in green (including fish and other aquatic animals), and Animal Output in orange (including livestock raised for meat, milk, eggs, and other animal-based products) from 1960 to 2019. The graphs are normalized so that the output of each country adds up to 100%.

When comparing Norway, Brazil, and India, there are significant differences. Crop Output represents the majority of Brazil's output (approximately 55%-65%) and India's output (approximately 65%-85%), while Norway's Crop Output has declined from around 35% to under 10% in recent decades. Animal Output has remained relatively steady in India (around 20%) and has slightly increased in Brazil (from 30% to 40%). However, Animal Output in Norway has decreased significantly from 65% to 20% in recent decades.

In Norway, Fish Output has replaced Animal Output, increasing from approximately 0% in 1970 to around 70% currently. Fish Output in Brazil and India represents less than 5% of their respective outputs, but has slightly increased in recent years.

In conclusion, different countries have different primary outputs based on their location. While Norway focuses on Fish Output, India and Brazil primarily produce Crop and Animal Output.

```python
# Method 3
obj.correlate_quantities()
```

After analysing the correlations between the different columns/variables of our dataset, we make the following observations:
* The "_quantity" columns are all correlated exclusively positively correlated which means that when a quantity increases, the other quantities will also increase. When a quantity decreases, the other quantities also decrease.

* The "_quantity" columns are correlated pretty high (>0.7) which means that the quantities have strong linear relationships to each other. Only the 'fish_output_quantity' has a generally weaker linear relationship to the other quantities. 

* There are 2 pairs of quantities that are perfectly, positively linear correlated (corr = 1): (ag land quantity and cropland quantity) & (crop output quantity and output quantity) which means that if one variable increases, the other variable also increases in a linear fashion.

* The lower correlation (0.66) between 'labor_quantity' and 'fish_output_quantity', compared to labor's relation to 'crop_output_quantity' and 'animal_output_quantity' (0.91 and 0.85 respectively), leads us to believe that fishing output is less dependent on labor than crops and animal output are.

* The lowest correlation (0.61-0.0.64) is between the column 'fish_output_quantity' and columns 'pasture_quantity', 'cropland_quantity', 'ag_land_quality'. This is to be expected as those variables have little to do with fishing:
    * pasture (land covered with grass and other low plants suitable for grazing animals)
    * cropland (land on which crops are grown) 
    * ag_land (agricultural land, the share of land area that is arable, under permanent crops, and under permanent pastures)
